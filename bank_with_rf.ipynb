```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("D:\kanb.csv")
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
X=pd.get_dummies(df,columns=['job','marital','education','default','housing','loan','contact','month','day_of_week','poutcome'],drop_first=True)
```

```python
X
```

```python
Y=pd.DataFrame(np.where(Y=='yes',1,0),columns=['y'])
```

```python
Y
```

```python
X
```

```python
X.drop(columns=['y'],inplace=True)
```

```python

import numpy as np
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)

print("Decision Tree accuracy",accuracy_score(y_test,y_pred))
```

```python

from sklearn.ensemble import RandomForestClassifier
```

```python

rf = RandomForestClassifier(n_estimators=500, random_state=42)
rf.fit(X_train, y_train)
```

# Random Forest without hyperparameter tuning

```python
y_pred = rf.predict(X_test)

print("Random forest accuracy",accuracy_score(y_test,y_pred))
```

```python
y_pred_train = rf.predict(X_train)

print("Random forest accuracy",accuracy_score(y_train,y_pred_train))
```

```python
# Number of trees in random forest
n_estimators = [20,60,100,120]

# Number of features to consider at every split
max_features = [0.2,0.6,1.0]

# Maximum number of levels in tree
max_depth = [2,8,None]

# Number of samples
max_samples = [0.5,0.75,1.0]

# 108 diff random forest train
```

```python
rf_grid.fit(X_train,y_train)
```

# Hyper Parameter tuning using random search cv

```python
# Number of trees in random forest
n_estimators = [20,60,100,120]

# Number of features to consider at every split
max_features = [0.2,0.6,1.0]

# Maximum number of levels in tree
max_depth = [2,8,None]

# Number of samples
max_samples = [0.5,0.75,1.0]

# Bootstrap samples
bootstrap = [True,False]

# Minimum number of samples required to split a node
min_samples_split = [2, 5]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
              'max_samples':max_samples,
              'bootstrap':bootstrap,
              'min_samples_split':min_samples_split,
              'min_samples_leaf':min_samples_leaf
             }
print(param_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV

rf_grid = RandomizedSearchCV(estimator = rf, 
                       param_distributions = param_grid, 
                       cv = 5, 
                       verbose=2, 
                       n_jobs = -1)
```

```python
rf_grid.fit(X_train,y_train)
```

```python
rf_grid.best_params_
```

```python
rf_grid.best_score_
```

```python
y_pred = rf_grid.predict(X_test)

print("Random forest accuracy",accuracy_score(y_test,y_pred))
```

```python
y_pred_train = rf.predict(X_train)

print("Random forest accuracy",accuracy_score(y_train,y_pred_train))
```
